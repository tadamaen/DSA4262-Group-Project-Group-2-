### **Project Introduction:**

Depression is a widespread mental health condition that goes beyond temporary sadness, affecting thoughts, emotions, behavior and physical well-being. It arises from a complex interplay of biological, psychological and social factors, with prolonged stress and lack of support increasing vulnerability. Globally, about 3.8% of the population, roughly 280 million people, experience depression, while rates in Singapore are notably higher at 6 to 7%. Young adults aged 15 to 29 and women are among the most affected groups.

Depression is deeply concerning because of its wide-ranging impact on individuals and society. It is the leading cause of disability worldwide, diminishing quality of life and functioning across personal, academic and social domains. Depression significantly increases the risk of suicide, particularly among young people, and is closely linked to physical health problems such as cardiovascular disease and other chronic conditions. On a broader scale, it imposes a heavy economic burden through reduced productivity and rising healthcare costs. These factors highlight why addressing depression is not only a matter of personal well-being but also a critical public health and societal priority.

### **Moving Into Singapore Context:**

Building on the broader societal impact of depression, our focus narrows to individuals at the National University of Singapore (NUS), particularly young people in their late teens to early adulthood. This demographic is especially vulnerable as young adults face unique pressures from academic demands, social transitions and the challenges of early independence. With depression rates in Singapore already higher than the global average, the need to address mental health among university students becomes even more pressing.

Despite the availability of counseling services, University Healthcare Centre (UHC) counselors encounter difficulties in providing continuous and timely support. Unavoidable gaps between counseling sessions can leave students without adequate monitoring, leading to delayed detection of worsening symptoms. These lapses increase the risk of relapse or more severe depressive episodes, underscoring the urgency of finding innovative solutions to bridge these support gaps and ensure sustained care for vulnerable students.

### **Data Cleaning And Preprocessing**

#### **Data Cleaning On Dataset 1 - IHME**

**Step 1: Removal Of Unnecessary Rows**

In this project, we are focusing specifically on individuals in Singapore whose reported cause is depressive disorders. To ensure a meaningful analysis, we will examine data across a 10-year horizon, filtering for records between 2014 and 2023 while excluding earlier years. This time frame captures the most recent trends and provides a balanced view of how depression has evolved in the country.

Additionally, our analysis will rely on percentage-based metrics rather than absolute numbers. Since Singapore's population has steadily grown over time, raw counts of depressive disorder cases would naturally increase and may not accurately reflect changes in prevalence. By using percentages, we obtain a fairer comparison that highlights the true proportion of people experiencing depression, offering clearer insights into the scale and impact of the condition across the population.

```python
# Step 1: Removal Of Unnecessary Rows
ihme = ihme[(ihme["location"] == "Singapore") &
            (ihme["cause"] == "Depressive disorders") &
            (ihme['year'] > 2013) & (ihme['year'] <= 2023) &
            (ihme['metric'] == 'Percent')]
```

**Step 2: Renaming Of Columns To More Readable Formats**

To improve clarity and consistency in the dataset, we will standardize the column names by renaming them into more readable formats. Instead of using underscores or technical abbreviations, each column will be written in camel case with spaces, such as changing `population_group` to Population Group and `metric` to Metric. This approach makes the dataset more intuitive and easier to interpret, especially for readers who may not be familiar with raw data structures.

```python
# Step 2: Renaming Of Columns
ihme = ihme.rename(columns={"population_group": "Population Group",
                            "measure": "Measure",
                            "location": "Location",
                            "sex": "Gender",
                            "age": "Age Range",
                            "cause": "Cause",
                            "metric" : "Metric",
                            "year": "Year",
                            "val": "Percentage Value",
                            "upper": "Upper Bound Percentage",
                            "lower": "Lower Bound Percentage"})
```

**Step 3: Removing Unnecessary Or Irrelavant Columns**

After filtering the dataset to focus only on Singapore, depressive disorders and percentage metrics, several columns become redundant and no longer add value to the analysis. For instance, the **Location** column contains only one value of Singapore, the **Cause** column is restricted to depressive disorders and the **Metric** column consistently shows percentages. Similarly, the **Population Group** column holds only a single entry of "All Population". Since these columns provide no variation or additional insights after filtering, they are irrelevant to our study and will be exclude.


```python
# Step 3: Removing Unnecessary And Irrelavant Columns
ihme = ihme.drop(columns=["Location", "Measure", "Cause", "Metric", "Population Group"])
ihme.head()
```

**Step 4: Checking For Any Missing/Null Values In Dataset**

To ensure the dataset is complete and reliable, the next step is to check for any missing or null values. Missing entries can compromise the accuracy of both exploratory analysis and predictive modeling, so it is important to identify them early. If null values are detected, we can address them in two main ways depending on the proportion of missing values.

If the dataset has only a small proportion of missing entries, removing the affected rows may be the simplest solution without significantly impacting the overall analysis. However, if missing values are more widespread, imputation techniques such as replacing them with the mean, median or mode of the column can be used to preserve the dataset's integrity.

```python
# Step 4: Checking For Missing Or Null Values
null_counts = ihme.isnull().sum()
print(null_counts)
```

Since it is confirmed that there are no missing or null values in any of the columns, the dataset is complete and requires no further treatment such as row removal or imputation. With data integrity ensured, we can now proceed to the next stage of the cleaning process.

**Step 5: Converting Values To Percentages**

Right now, the values in the **Percentage Value**,
**Upper Bound Percentage**, and **Lower Bound Percentage** columns are expressed in decimal form, which can be misleading given the column names. To make the dataset more consistent and reader-friendly, we should convert these decimals into actual percentages. This will be done by multiplying each value in the three columns by 100, ensuring that the data aligns with the column labels and is easier to interpret during exploratory data analysis.


```python
# Step 5: Converting Values Into Percentages
ihme["Percentage Value"] = ihme["Percentage Value"] * 100
ihme["Upper Bound Percentage"] = ihme["Upper Bound Percentage"] * 100
ihme["Lower Bound Percentage"] = ihme["Lower Bound Percentage"] * 100
```

**Step 6: Regrouping The Age Range Column**

At present, the Age Range column contains multiple narrow intervals such as "15-19 years", "20-24 years" and "25-29 years".Since most NUS students fall between 15 and 24 years old, we will reclassify the age ranges into broader 10-year intervals. This reduces the number of categories, making the dataset easier to interpret and more aligned with the target population.

For the associated values in **Percentage Value**, **Upper Bound Percentage** and **Lower Bound Percentage**, we will calculate averages within each new age group. While a more precise method would involve weighting percentages by absolute case counts, simplifying the analysis by averaging percentages (grouped by gender and year) provides a clear and manageable approach. This adjustment ensures that the dataset remains focused, interpretable and relevant to the study of depression among university-aged individuals in Singapore.

```python
# Step 6: Regrouping The Age Range Column
def reclassify_age(age):
    if age in ["15-19 years", "20-24 years"]:
        return "15-24 years"
    elif age in ["25-29 years", "30-34 years"]:
        return "25-34 years"
    elif age in ["35-39 years", "40-44 years"]:
        return "35-44 years"
    elif age in ["45-49 years", "50-54 years"]:
        return "45-54 years"
    elif age in ["55-59 years", "60-64 years"]:
        return "55-64 years"
    elif age in ["65-69 years", "70-74 years"]:
        return "65-74 years"
    elif age in ["75-79 years", "80-84 years"]:
        return "75-84 years"
    else:
        return "> 85 years"

ihme["Age Range"] = ihme["Age Range"].apply(reclassify_age)

ihme = ihme.groupby(["Gender", "Age Range", "Year"], as_index=False).agg({
    "Percentage Value": "mean",
    "Upper Bound Percentage": "mean",
    "Lower Bound Percentage": "mean"})

ihme.head()
```

From the results, we observed that certain columns which should not contain null values still have missing entries. For example, the `waiting_time` column has 78 null values, even though it is expected to have none. This raises the need to address missing data either by removing affected rows or applying imputation techniques such as using the mode or median for numerical variables.

To determine the most appropriate approach, we will first calculate the proportion of rows that contain at least one null value across all columns where nulls are not permissible. If this proportion is relatively low (< 20%), we will likely remove those rows to preserve data integrity. Conversely, if the proportion is higher (≥ 20%), imputation methods will be considered to retain sufficient data for analysis.

```python
# Finding The Proportion Of Rows With At Least 1 NA Value

# Define columns that should not have null values
required_columns = ["Waiting Time", "Prior Treatment", "Number Of Sessions",
                    "Session Frequency", "PHQ-9 Score (Session 1)", "Initial PHQ Depression Level"]

# Check rows with at least one null in required columns
rows_with_nulls = cbt[required_columns].isnull().any(axis=1)

# Calculate proportion
proportion = rows_with_nulls.mean() * 100
print(f"Proportion Of Rows With At Least One Null Value: {proportion:.2f}%")
```

Since the proportion of rows containing null values in required columns is approximately 10%, this represents a relatively small fraction of the dataset. To preserve data accuracy while maintaining a robust sample size, we will proceed by removing these rows. This approach ensures that the remaining 90% of the dataset is complete and reliable for subsequent analysis.

```python
# Drop rows with nulls in required columns
cbt = cbt.dropna(subset = required_columns)

null_counts = cbt.isnull().sum()
print(null_counts)
```

After removing the rows with null values in the required columns, we confirm that these critical fields are now complete with no missing entries. While some null values remain in the PHQ-9 Score columns for Sessions 2 through 10, these are expected since not all patients attend the full set of sessions. #### **Visualization 1: Determining Proportion Of Depression Among Different Age Groups In Singapore Over Time**

#### **Visualization 3: Year-On-Year Percentage Change In Depression Rates In Singapore (Age Group 15 - 24)**

At a more granular level, our analysis will now focus specifically on individuals within the 15 - 24 years age group. Using the IHME dataset, we can compute the year-on-year (YoY) percentage change in depression prevalence from 2014 to 2023. This calculation provides insight into how depression rates evolve annually, highlighting both gradual shifts and sudden spikes. The YoY change is determined by comparing the current year's rate with the previous year's rate, using the formula:

**YOY Change (%) = (Rate(present) - Rate(1 year ago)) / Rate(1 year ago)**

This metric captures the relative increase or decrease in prevalence from one year to the next. By plotting these changes on a line graph, we can visualize the trajectory of depression rates over time. Significant anomalies, such as unusually high increases, will be highlighted in red to emphasize major shifts in mental health trends among young individuals. This approach allows us to identify critical years where depression prevalence surged, offering valuable context for understanding underlying causes and poteial interventions.



```python
# Visualization 3: Year-On-Year Percentage Change In Depression Rates In Singapore (15 to 24)

ihme_subset = ihme[(ihme["Gender"] == "Both") &
                   (ihme["Age Range"] == "15-24 years")].copy()

ihme_subset["Percentage Change"] = ihme_subset.groupby("Age Range")["Percentage Value"].pct_change() * 100


# Plot line graph
fig, ax = plt.subplots(figsize=(12, 7))
sns.lineplot(data=ihme_subset, x="Year", y="Percentage Change", marker="o", markersize=8,
             errorbar=None, ax=ax)

# Dynamic offset for labels
y_offset = (ihme_subset["Percentage Change"].max() - ihme_subset["Percentage Change"].min()) * 0.02

for x, y in zip(ihme_subset["Year"], ihme_subset["Percentage Change"]):
    if not np.isfinite(y):
        continue
    x_adjust = 0
    if x == 2019:
        x_adjust = -0.15
    elif x == 2021:
        x_adjust = 0.15
    # Highlight specific years
    if x in [2020, 2021]:
        color = "red"
        fontweight = "bold"
    elif x == 2022:
        color = "green"
        fontweight = "bold"
    else:
        color = "black"
        fontweight = "normal"

    ax.text(x + x_adjust, y + y_offset if y >= 0 else y - y_offset, f"{y:.2f}", ha="center",
            va="bottom" if y >= 0 else "top", fontsize=9, color=color, fontweight=fontweight)

ax.set_title("Year-On-Year Percentage Change in Depression Rates in Singapore (Age 15 – 24)",
             fontweight="bold", fontsize=13, pad=15)
ax.set_xlabel("Year", fontweight="bold")
ax.set_ylabel("Year-On-Year Percentage Change (%)", fontweight="bold", fontsize=10)
ax.axhline(0, linestyle="--", linewidth=1)
plt.ylim(ihme_subset["Percentage Change"].min() - 2, ihme_subset["Percentage Change"].max() + 2)
plt.tight_layout()
plt.show()
```

**Key Observations:**

1) In 2020 and 2021, there was a sharp rise in depression rates among individuals aged 15 to 24, with increases of nearly 9% and 6.5% respectively. These years stand out as anomalies compared to the relatively stable and modest changes observed in earlier years.

2) In 2022, the trend reversed with a notable decline of about 4%. This drop suggests a partial correction following the pandemic-related spikes, though the decrease did not fully offset the earlier increases.

3) By 2023, the rates rose again by approximately 1.5%. While this increase was smaller than the pandemic-era surges, it indicates that the overall upward trajectory in depression prevalence continues.

**Possible Insights:**

- The decline in 2022 suggests some stabilization as restrictions eased and social activities resumed, but the persistence of elevated rates highlights that the pandemic's effects were not fully reversed.

##### **Metric 3: Average Change In PHQ-9 Scores Per Counselling Session**

For this metric, we need to create an additional column called `PHQ-9 Score Rate Change` for the calculation. An example of the calculation is shown below: 

Patient A: Attended 5 Counselling Sessions

*   `PHQ-9 Score (Session 1)`: 16
*   `PHQ-9 Score (Session 2)`: 14
*   `PHQ-9 Score (Session 3)`: 13
*   `PHQ-9 Score (Session 4)`: 13
*   `PHQ-9 Score (Session 5)`: 11

We will take the first and the last session scores, which means 
`PHQ-9 Score (Session 1)` of 16 and `PHQ-9 Score (Session 5)` of 11 and divide by the number of sessions, which is 5. In this example, Average Change In PHQ-9 Score will be (11-16) / 5 = **-1**. We will apply the same logic to all the patients in the CBT dataset and calculate the average of all the values.


```python
phq_columns = ["PHQ-9 Score (Session 1)", "PHQ-9 Score (Session 2)", "PHQ-9 Score (Session 3)",
               "PHQ-9 Score (Session 4)", "PHQ-9 Score (Session 5)", "PHQ-9 Score (Session 6)",
               "PHQ-9 Score (Session 7)", "PHQ-9 Score (Session 8)", "PHQ-9 Score (Session 9)",
               "PHQ-9 Score (Session 10)"]

# Function to calculate average change per patient
def avg_change(row):
    # Drop missing values for this patient
    scores = row[phq_columns].dropna()
    if len(scores) < 2:
        return None 
    first_score = scores.iloc[0]
    last_score = scores.iloc[-1]
    num_sessions = len(scores)
    return (last_score - first_score) / num_sessions

# Apply function to each patient
cbt["PHQ-9 Score Rate Change"] = cbt.apply(avg_change, axis=1)

# Calculate overall average change across all patients
overall_avg_change = cbt["PHQ-9 Score Rate Change"].mean()
print("Average Change In PHQ-9 Score Per Counselling Session:", round(overall_avg_change, 2))
```

The analysis shows that the average change in PHQ-9 score per counselling session prior to the intervention is **approximately -0.66**. This indicates that on average, patients in the CBT dataset experience a reduction of approximately 0.66 points in their PHQ-9 score with each successive session. In practical terms, this reflects a gradual decrease in depression severity as patients continue attending sessions. While the pace of improvement is relatively modest, the consistent downward trend demonstrates that counselling is contributing positively to symptom reduction over time.
